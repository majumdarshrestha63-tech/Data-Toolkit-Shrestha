6. Python lists and NumPy arrays are both used to store collections of data, but they differ significantly in their characteristics and use cases.

  Data Type Homogeneity:

Python List: Can store heterogeneous data types, meaning elements within a single list can be of different types (eg  integers, floats, strings, or even other lists).

NumPy Array: Stores homogeneous data types, requiring all elements within an array to be of the same type. This allows for more efficient storage and computation.

  Performance:

Python List: Generally less efficient for numerical operations on large datasets due to the overhead of storing and managing heterogeneous data types. Operations often require explicit loops.

NumPy Array: Optimized for numerical computations, offering significantly better performance for large datasets. This is achieved through vectorized operations, which are implemented in C and Fortran for speed.

  
  Memory Efficiency:

Python List: Less memory efficient, especially for large datasets, as each element is a full Python object with its own type information and reference count.

NumPy Array: More memory efficient for large numerical datasets because it stores data in contiguous memory blocks and avoids the overhead of individual Python objects for each element.


 Functionality:


Python List: Provides general-purpose list manipulation methods (eg, append(), insert(), remove()).

NumPy Array: Offers a rich set of mathematical functions and operations optimized for array manipulation, including advanced indexing, broadcasting, and linear algebra routines.

 Use Cases:


Python List: Suitable for general-purpose tasks where flexibility in data types and dynamic resizing are important, such as storing diverse collections of objects or managing queues.

NumPy Array: Ideal for numerical computing, scientific applications, and data analysis involving large datasets where performance and memory efficiency are crucial, such as image processing, machine learning, and simulations.

9. Matplotlib and Plotly are both powerful Python visualization libraries, but they serve different purposes and are built with different strengths in mind. The differences are as follows:-

   Static vs Interactive Plots

   Matplotlib- Primarily designed for static, publication-quality plots. Plots are images (PNG, PDF, SVG). Interaction is minimal unless using additional tools.

   Plotly- Built for interactive visualizations (zoom, hover tooltips, sliders). Generates HTML-based charts ideal for dashboards and web apps.


   Ease of Use

   Matplotlib- Low-level library → gives full control but requires more code. Often need to manually manage labels, ticks, layout, etc.

   Plotly- Higher-level interactions come built-in. Complex visuals (3D plots, maps, subplots) can be created with fewer lines of code.


   Appearance and Styling

   Matplotlib- Default style is simple and classic. Highly customizable but requires manual styling. Best for scientific publications (e.g journals).

   Plotly- Modern, colorful, and attractive by default. Interactive legends, animations, and hover details included automatically.


   Use Cases

   When to use Matplotlib- Academic research, Static charts for reports and papers , Precise control over every visual element, Integration with NumPy/SciPy workflow ,Reproducible static plots for Jupyter notebooks
   
   
   When to use Plotly- Dashboards (especially with Plotly Dash), Web-based visualizations, Interactive data exploration, Presentations where user interaction is helpful, 3D plots, maps, time-series with zoom/hover


   Supported Plot Types

   Matplotlib: 2D plots (line, bar, histogram, scatter), Limited 3D support (via mpl_toolkits.mplot3d), Custom shapes, subplots, and annotations

   Plotly: Everything Matplotlib can do +,Interactive 3D plots, Geographical maps, Animations, Interactive heatmaps and treemaps, Financial charts (candlesticks, OHLC)


   Performance

   Matplotlib: Faster for large static datasets. Ideal for millions of points when interaction isn't required.

   Plotly: Interactive rendering can be slower for extremely large datasets. Uses JavaScript under the hood.


   Output Formats

   Matplotlib: PNG, JPG, SVG, PDF, Good for print media

   Plotly: HTML, JSON, Embeddable interactive visuals for web apps, Also supports static export (via Kaleido)

18. The advanced features of NumPy are as follows:-
    
    Broadcasting: This is a powerful mechanism that allows NumPy to work with arrays of different shapes during arithmetic operations by "broadcasting" the smaller array across the larger one. This avoids the need for explicit loops and memory-intensive data duplication, making code more efficient and concise.
    
    Advanced Indexing and Slicing: Beyond basic slicing, NumPy supports advanced indexing with lists or arrays of integers to select specific elements and using boolean arrays to filter elements based on a condition.
    
    Vectorized Operations: NumPy performs mathematical operations on entire arrays at once, known as vectorization, which is significantly faster than using traditional Python loops.
    
    Universal Functions (ufuncs): These are functions that operate on ndarrays element-wise, such as np.sin(), np.cos(), and np.exp(), providing a fast and efficient way to perform calculations.

    Specialized modules and functionalities

    Linear Algebra: NumPy includes a comprehensive linear algebra module that performs operations like matrix multiplication, finding the determinant, inverse, rank, and trace.
    
    Fourier Transforms: It has built-in capabilities to perform Fourier transforms on arrays.
    
    Array Manipulation: Advanced array manipulation functions include stacking arrays vertically (vstack) or horizontally (hstack), and splitting arrays along different axes.
    
    Random Number Generation: NumPy provides a random module for generating random numbers for a wide range of distributions and scenarios.
    
    Statistical Functions: It offers a wide array of functions for statistical analysis, such as calculating the mean, median, variance, and finding the maximum or minimum values in an array.
    
    Data Type Handling: NumPy arrays are homogeneous, meaning they store elements of a single data type, which improves performance and memory efficiency. It also provides tools for working with a variety of data types, including datetime64 for date and time functionality.

22. Common use cases for Seaborn are as follows:-
    
    Exploring Relationships between Variables:

    Scatter plots (sns.scatterplot): To visualize the relationship between two continuous variables and identify trends, clusters, or outliers.

    Line plots (sns.lineplot): To show trends over time or across different categories, especially for continuous data.

    Pair plots (sns.pairplot): To visualize the relationships between all pairs of variables in a dataset, as well as the distributions of individual variables.


    Visualizing Data Distributions:

    Histograms (sns.histplot): To display the frequency distribution of continuous data, revealing its shape, center, and spread.

    Kernel Density Estimates (KDEs) (sns.kdeplot): To visualize the probability density function of a continuous variable, providing a smoothed representation of its distribution.

    Box plots (sns.boxplot): To display the distribution of data and identify outliers, particularly useful for comparing distributions across different categories.

    Violin plots (sns.violinplot): To combine the features of box plots and KDEs, showing both the distribution and central tendency of data across categories.

    Comparing Categorical Data:

    Bar plots (sns.barplot): To visualize counts, frequencies, percentages, or averages of categorical variables.

    Count plots (sns.countplot): To display the number of occurrences of each category in a categorical variable.

    Cat plots (sns.catplot): A versatile function to create various types of categorical plots, such as box plots, violin plots, and bar plots, with options for faceting and grouping.

    Visualizing Data in Grid Formats:

    Heatmaps (sns.heatmap): To visualize data in a grid format, often used to display correlation matrices or patterns in tabular data.

    Enhancing Plot Aesthetics:
    
    Seaborn provides built-in themes, color palettes, and functions to easily create visually appealing plots with minimal code, improving the overall presentation of data insights.


    Facilitating Exploratory Data Analysis (EDA):

    Seaborn's dataset-oriented API and high-level functions make it easy to quickly generate a variety of plots, aiding in the initial exploration and understanding of data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 2, 3, 2]
}

df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
